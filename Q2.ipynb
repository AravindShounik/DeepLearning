```python
# Importing Libraries

import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating data samples from two Gaussian distributions

D1 = torch.normal(0.0, 1.0, [500, 2])
D2 = torch.normal(3.0, 1.0, [500, 2])
```

```python
# Plotting the data samples

plt.scatter(D1[:, 0], D1[:, 1], color='red')
plt.scatter(D2[:, 0], D2[:, 1], color='blue')
```

```python
X = torch.empty(1000, D1.shape[1] + 1).fill_(1)
# print(X.size())
X[::2,:2] = D1
X[1::2,:2] = D2
# print(D1)
# print(X)

# Creating the labels
Y = torch.empty(1000, 1)
Y[::2,:2] = 1.0
Y[1::2,:2] = -1.0
```

```python
def loss(X,Y,W):
    return torch.mean((torch.matmul(X,W) - Y)**4)
```

```python
def grad(X,Y,W):
    return torch.mean(4*(torch.matmul(X,W) - Y)**3 * X, dim=0)
```

```python
lr = 0.01
W = torch.zeros(3,1)
no_iterations = 1000
for i in range(no_iterations):
    W = W - lr * grad(X,Y,W).reshape(3,1)
    print(loss(X,Y,W))

print(X.size(),W.size())
```

```python
# Plotting the data samples
a = torch.matmul(X,W)
for x in range(X.size(0)):
    if a[x] >= 0:
        plt.scatter(X[x, 0], X[x, 1], color='red')
    else:
        plt.scatter(X[x, 0], X[x, 1], color='blue')
```
