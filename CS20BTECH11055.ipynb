```python
# Importing Libraries

import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
# Creating a Linearly Separable Dataset with 2 classes and 2 features

def create_dataset(n_samples, gamma):
    X = torch.zeros(n_samples, 3)
    for i in range(n_samples):
        if i > n_samples/2:
            X[i,0] = np.random.uniform(0,20)
            X[i,1] = X[i,0] + np.random.uniform(0,3)
            X[i,2] = 1
        else:
            X[i,0] = np.random.uniform(0,20)
            X[i,1] = X[i,0] + gamma - np.random.uniform(0,3)
            X[i,2] = 1
    # Creating Labels for the data set
    Y = torch.zeros(n_samples, 1)
    for i in range(n_samples):
        if i > n_samples/2:
            Y[i] = 1
        else:
            Y[i] = -1
    return X, Y
```

```python
# Perceptron Training Algorithm

def train_perceptron(x, y, w, nb_epochs_max):  
  count = 0
  for e in range(nb_epochs_max):
    nb_changes = 0
    for i in range(x.size(0)):
      if x[i].dot(w) * y[i] <= 0:
        w = w + y[i] * x[i]
        count+=1
        nb_changes = nb_changes + 1
    #print('Number of changes: %d'%nb_changes) # prints the number of changes every epoch
    if nb_changes == 0:
        print('Early stopping at epoch number %d',e) 
        break
  print('Number of changes: %d'%count)  # prints the number of chages in the end  
  return w
```

```python
D1, L1 = create_dataset(1000, 10)
D2, L2 = create_dataset(1000, 11)
D3, L3 = create_dataset(1000, 12)
D4, L4 = create_dataset(1000, 13)
D5, L5 = create_dataset(1000, 14)
```

```python
# Defining loss function

def loss_function(x, y, w):
    return torch.mean((torch.matmul(x, w) - y) ** 4,dim=0)
```

```python
# Defining Gradient 

def gradient(x, y, w):
    # print("Hi", x.size())
    out = 4 * ( torch.matmul(x,w) - y ) ** 3 * x
    output = torch.zeros(x.size(1),1)
    for i in range(x.size(1)):
        for j in range(x.size(0)):
            output[i,0] += out[j,i]
        output[i,0] = output[i,0]/x.size(0)
    return output
```

```python
# Defining Accuracy Function

def accuracy(x, y, w):
    count = 0
    a = torch.matmul(x,w)
    for i in range(x.size(0)):
        if a[i]*y[i] > 0:
            count+=1
    return count/x.size(0)
```
